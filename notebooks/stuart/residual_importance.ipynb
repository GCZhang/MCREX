# Variance Reduction with a Residual Importance Map

An importance map, $\phi$, may be used to describe how important a particular region of the problem is to Monte Carlo calculation. Using this importance map we can the prescribe a variance reduction scheme that favors regions of high importance while attempting to minimize the work performed in regions of low importance. To do this, we need to prescribe the behaviour of a Monte Carlo history is it moves between regions of different importance.

### Russian Roulette

First consider when a history moves to a region of high importance to a region of lower importance in a transition $i \rightarrow j$ such that $|\phi_i| > |\phi_j|$. In this case with then play the roulette game based on the survival parameter $\theta = \frac{|\phi_i|}{|\phi_j|}$ such that given a random number $\xi$ if:
\begin{equation}
    \xi > \frac{1}{\theta}
\end{equation}
the history is terminated and if:
\begin{equation}
    \xi \geq \frac{1}{\theta}
\end{equation}
then the history survives with weight $W = \theta W$.

```python
import random 

def roulette( phi_i, phi_j ):
    theta = abs(phi_i/phi_j)
    rn = random.random()
    kill = rn > 1.0 / theta
    return kill
```

### Splitting

Next we consider moving from a region of low importance to a region of higher importance for a transition $i \rightarrow j$ such that $|\phi_i| < |\phi_j|$. In this case we split the history into a number of histories with smaller weights such that if $\phi_i$ and $\phi_j$ are integers then the number of new histories is:
\begin{equation}
    N_{split} = \frac{|\phi_j|}{|\phi_i|}
\end{equation}
and their weight is $W = \frac{W}{N_{split}}$. We modify this slightly for when these numbers are not integers.

```python
import math

# Tom Booth variance reduction paper - sample splitting to preserve expected value.

def split( phi_i, phi_j ):
    n_split = int( math.floor(abs(phi_j/phi_i)) )
    return n_split
```

### Non-analog Sampling Procedure

We can now write a non-analog sampling procedure for the Monte Carlo solution of a linear system. As input we have the right hand side, b, the CDF for the right hand side, and CDF and transition weight matrix for the linear operator, a weight cutoff, the importance map, a roulette parameter, and a running sum of the mean tally and variance tally.

```python
import numpy
import mc_tools

def doOneHistory( b, b_cdf, C, W, w_f, phi, x, sigma):
    
    # Get the initial state of the history.
    state = mc_tools.sampleSourceCDF( random.random(), b_cdf )
    starting_weight = numpy.linalg.norm( b, 1 )
    
    # Make a set of stacks for histories and current weights created from splitting. 
    history_stack = []
    weight_stack = []
    history_stack.append(state)
    weight_stack.append( starting_weight*numpy.sign(b[state]) )
    
    # This stack tracks the effective 'weight' of a history. This is
    # different than the weight used in the calculation of the linear system.
    hw_stack = []
    hw_stack.append( 1.0 )

    # Initialize an empty tally vector for this history
    size = len( x )
    tally = numpy.zeros(size)
    
    # Run all histories in the stack.
    while len(history_stack) > 0:

        # Get the history on the top of the stack.
        state = history_stack.pop()
        weight = weight_stack.pop()
        hw = hw_stack.pop()
        
        # Run history on the top of the stack until the cutoff or roulette/split.
        while abs(weight) > w_f:

            # Process a transition
            tally[state] = tally[state] + hw*weight
            new_state = mc_tools.sampleMatrixCDF( state, random.random(), C )
            weight = weight * W[state][new_state]

            # Roulette if the importance decreased
            if ( abs(phi[new_state]) < abs(phi[state]) ):

                # Set the weight to zero to kill the history.
                if ( roulette(phi[state],phi[new_state]) ):
                    weight = 0.0

                # If it survived set the weight to the survival weight.
                else:
                    hw = hw*abs(phi[state]/phi[new_state])

            # Split if above weight window
            elif ( abs(phi[new_state]) > abs(phi[state]) ):

                # Compute the number of new histories and push them
                # onto the stack.
                n_split = split( phi[state], phi[new_state] )
                w_split = hw / n_split
                for j in xrange(n_split):
                    history_stack.append(new_state)
                    weight_stack.append(weight)
                    hw_stack.append(w_split)

                # End the current history and instead process the new histories
                # from the split.
                weight = 0.0
                
            # Update the state.
            state = new_state
        
    # Combine the tally results for the mean and variance.
    for i in xrange(size):
        x[i] = x[i] + tally[i]
        sigma[i] = sigma[i] + tally[i]*tally[i]

    # Return the mean and variance estimates.
    return (x, sigma)
```

### Monte Carlo Linear Solver

Using the modified sampling procedure we can now write a non-analog Monte Carlo solver for linear systems using the adjoint method. The inputs are the linear operator, the right hand side, the residual, the weight cutoff, and the number of histories.

```python
import mc_data

def nonAnalogMC( A, b, phi, w_c, np ):
    
    # Make the standard Monte Carlo data structures.
    H, C, W = mc_data.makeMonteCarloHCW( A )
    b_cdf, starting_weight = mc_data.makeSourceCDF( b )
    w_f = w_c * starting_weight
    
    # Solve the problem with non-analog Monte Carlo.
    x = numpy.zeros( len(b) )
    sigma = numpy.zeros( len(b) )
    for i in xrange(np):
        x, sigma = doOneHistory( b, b_cdf, C, W, w_f, phi, x, sigma )

    # Normalize the mean and variance tallies.    
    for i in xrange( len(b) ):
        x[i] = x[i] / np
        sigma[i] = (sigma[i] / np - x[i]*x[i])/np
    return (x, sigma)
```

## Model Problem

We solve the following linear problem on one-dimension:
\begin{equation}
    \nabla^2 \mathbf{u} - D \mathbf{u} = \mathbf{f}
\end{equation}
with $\mathbf{u} \in \mathbf{R}^N$ the solution variable, $\mathbf{f} \in \mathbb{R}^N$ the forcing term, and D a scalar. We discretize with a cell-centered finite difference with mesh size h:
\begin{equation}
    \frac{1}{h^2}(u_{i-1} - 2u_i + u_{i+1}) - D u_{i-1} = f_i
\end{equation}
The boundary conditions on the left are:
\begin{equation}
    \frac{1}{h^2}(u_{-1} - 2u_0 + u_{1}) -D u_0 = f_0
\end{equation}
On a cell-centered grid we have:

We set $u_{-1/2} = 0$ so that
\begin{equation}
\frac{1}{h^2}(-3u_0 + u_1) - D u_0= f_0
\end{equation}

This gives the following linear operator:
\begin{equation}
    \nabla^2 = \frac{1}{h^2}
    \begin{bmatrix}
        -3-D & 1 & & & & & \\
         1 & -2-D & 1 & & & & \\
           & 1 & -2-D & 1 & & & \\
           & & & \cdots & & & \\
           & & & & 1 & -2-D & 1 \\
           & & & & & 1 & -3-D \\
    \end{bmatrix}
\end{equation}

We will construct the diagonally-scaled operators on each grid next:

```python
grid_size = 100
h = 1.0
```

```python
import array_tools
import mc_data

D = 0.1

A = array_tools.makePoissonOperator( grid_size, h, D )
diag = array_tools.getInvDiag( A )

A = array_tools.leftScaleMatrix( A, diag )

H = mc_data.makeIterationMatrix( A )
eval_H, evec_H = numpy.linalg.eig(H)
spec_rad_H = max(sorted(abs(i) for i in eval_H))
print "rho(H):", spec_rad_H
```

### Forcing Term

Next make a forcing term. Choose a forcing term of unity to make it simple to define on each grid. We apply the diagonal scaling here as well. Set use_random to true to add some randomness to the right hand side.

```python
use_random = True
f = numpy.zeros( grid_size )
for i in xrange(grid_size):
    if use_random:
        f[i] = 1.0 + random.random() * random.random()
    else:
        f[i] = 1.0
f = array_tools.scaleVector( f, diag )
```

## Reference Solution

```python
ref = numpy.linalg.solve(A,f)
```

## Monte Carlo Solution

Set the weight cutoff and number of histories. Also set a random initial guess and compute the initial residual.

```python
w_c = 1.0e-2
np = 1000

x_0 = numpy.zeros( grid_size )
for i in xrange( grid_size ):
    x_0[i] = 1.0 + random.random()
    
r_0 = array_tools.computeResidual( A, x_0, f )
```

### Analog Monte Carlo

```python
import time
import mc_tools

x = numpy.zeros( grid_size )
sigma = numpy.zeros( grid_size )

start = time.time()
delta, sigma = mc_tools.monteCarloSolve( A, r_0, w_c, np)
end = time.time()
time_x = end - start
print "MC Time:", time_x

for i in xrange(grid_size):
    x[i] = x_0[i] + delta[i]
```

### Monte Carlo with Residual Importance Sampling

```python
import variance_reduction

x_ris = numpy.zeros( grid_size )
sigma_ris = numpy.zeros( grid_size )

start = time.time()
delta_ris, sigma_ris = nonAnalogMC( A, r_0, r_0, w_c, np )
end = time.time()
time_ris = end - start
print "MC RIS Time:", time_ris

for i in xrange(grid_size):
    x_ris[i] = x_0[i] + delta_ris[i]
```

## Analysis

Compute the solution residual and relative variance norm

```python
r = array_tools.computeResidual(A,x,f)
r_norm = numpy.linalg.norm(r,2)
f_norm = numpy.linalg.norm(f,2)
r_scale = numpy.zeros( grid_size )
for i in xrange( grid_size ):
    r_scale[i] = abs( r[i] / f[i] )
rel_var = numpy.zeros( grid_size )
for i in xrange(grid_size):
    rel_var[i] = 100.0 * sigma[i] / abs(x[i])
var_norm = numpy.linalg.norm(rel_var,2)

r_ris = array_tools.computeResidual(A,x_ris,f)
r_ris_norm = numpy.linalg.norm(r_ris,2)
r_ris_scale = numpy.zeros( grid_size )
for i in xrange( grid_size ):
    r_ris_scale[i] = abs( r_ris[i] / f[i] )
rel_var_ris = numpy.zeros( grid_size )
for i in xrange(grid_size):
    rel_var_ris[i] = 100.0 * sigma_ris[i] / abs(x_ris[i])
var_ris_norm = numpy.linalg.norm(rel_var_ris,2)

print "residual 2-norm:", r_norm / f_norm, r_ris_norm / f_norm
print "variance 2-norm:", var_norm, var_ris_norm
```

Compute the figure of merit based on the residual

```python
rfom = 1.0 / (time_x*r_norm*r_norm/(f_norm*f_norm))
rfom_ris = 1.0 / (time_ris*r_ris_norm*r_ris_norm/(f_norm*f_norm))
print "Relative Residual FOM:", rfom_ris / rfom
```

Plot the results

```python
%matplotlib inline
import matplotlib.pyplot as plt
from pylab import *

plot_grid = numpy.linspace(0,grid_size,grid_size)
figure(figsize=(10,8), dpi=100)
plot(plot_grid,ref,'k',label="Ref")
plot(plot_grid,x,'r',label="Analog MC")
plot(plot_grid,x_ris,'b',label="RIS MC")
title("Solution")
legend()
show() 
```

```python
figure(figsize=(10,8), dpi=100)
plot(plot_grid,r_scale,'r',label="Analog MC")
plot(plot_grid,r_ris_scale,'b',label="RIS MC")
title("Residual")
yscale('log')
legend()
show() 
```

Plot the relative variance

```python
figure(figsize=(10,8), dpi=100)
plot(plot_grid,rel_var,'r',label="Analog MC")
plot(plot_grid,rel_var_ris,'b',label="RIS MC")
title("Relative Variance")
yscale('log')
legend()
show() 
```

Compute the figure of merit based on the variance

```python
fom = numpy.zeros( grid_size )
fom_ris = numpy.zeros( grid_size )
for i in xrange(grid_size):
    fom[i] = 1.0 / (rel_var[i]*time_x)
    fom_ris[i] = 1.0 / (rel_var_ris[i]*time_ris)
    
print "Minimum FOM:", min(fom), min(fom_ris)
print "Maximum FOM:", max(fom), max(fom_ris)
print "FOM 1-norm:", sum(fom), sum(fom_ris)
```

Plot the figure of merit

```python
figure(figsize=(10,8), dpi=100)
plot(plot_grid,fom,'r',label="Analog MC")
plot(plot_grid,fom_ris,'b',label="RIS MC")
title("Figure of Merit")
yscale('log')
legend()
show() 
```
