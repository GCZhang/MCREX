# Three-Level Monte Carlo Scheme

This study analyzyes a three-level Monte Carlo scheme applied to a Poisson problem discretized by a cell-centered finite difference scheme.

## Multilevel Scheme

We explore here the basics of the multilevel Monte Carlo scheme when applied to the solution of linear systems. Consider the two level estimator of Giles:

\begin{equation}
   E(u) = E(u_2) + E(u_1-u_2) + E(u_0-u_1)
\end{equation}

Where $E(u_1)$ is the estimate on the coarse level and $E(u_0)$ is the estimate on the fine level. One can interpret this as correcting the coarse solution with a difference between the fine and coarse solution. We will then have 3 grids:

The key behind constructing the multilevel estimate is that separate sets of random walks are performed, one on grid. For the random walks not on the coarsest grid, two tallies are performed on both the coarse and fine grids with their difference computed to construct the estimator (i.e. $E(u_0-u_1)$). Let's make the grids:

```python
import numpy

grid_0_size = 40
h_0 = 1.0

grid_1_size = grid_0_size / 2
h_1 = h_0*2.0

grid_2_size = grid_1_size / 2
h_2 = h_1*2.0
```

## Model Problem

We solve the following linear problem on one-dimension:
\begin{equation}
    \nabla^2 \mathbf{u} - D \mathbf{u} = \mathbf{f}
\end{equation}
with $\mathbf{u} \in \mathbf{R}^N$ the solution variable, $\mathbf{f} \in \mathbb{R}^N$ the forcing term, and D a scalar. We discretize with a cell-centered finite difference with mesh size h:
\begin{equation}
    \frac{1}{h^2}(u_{i-1} - 2u_i + u_{i+1}) - D u_{i-1} = f_i
\end{equation}
The boundary conditions on the left are:
\begin{equation}
    \frac{1}{h^2}(u_{-1} - 2u_0 + u_{1}) -D u_0 = f_0
\end{equation}
On a cell-centered grid we have:

We set $u_{-1/2} = 0$ so that
\begin{equation}
\frac{1}{h^2}(-3u_0 + u_1) - D u_0= f_0
\end{equation}

This gives the following linear operator:
\begin{equation}
    \nabla^2 = \frac{1}{h^2}
    \begin{bmatrix}
        -3-D & 1 & & & & & \\
         1 & -2-D & 1 & & & & \\
           & 1 & -2-D & 1 & & & \\
           & & & \cdots & & & \\
           & & & & 1 & -2-D & 1 \\
           & & & & & 1 & -3-D \\
    \end{bmatrix}
\end{equation}

We will construct the diagonally-scaled operators on each grid next:

```python
import array_tools
import mc_data
D = 0.01

A_0 = array_tools.makePoissonOperator( grid_0_size, h_0, D )
A_1 = array_tools.makePoissonOperator( grid_1_size, h_1, D )
A_2 = array_tools.makePoissonOperator( grid_2_size, h_2, D )

diag_0 = array_tools.getInvDiag( A_0 )
diag_1 = array_tools.getInvDiag( A_1 )
diag_2 = array_tools.getInvDiag( A_2 )

A_0 = array_tools.leftScaleMatrix( A_0, diag_0 )
A_1 = array_tools.leftScaleMatrix( A_1, diag_1 )
A_2 = array_tools.leftScaleMatrix( A_2, diag_2 )
```

Next make a forcing term. Choose a forcing term of unity to make it simple to define on each grid. We apply the diagonal scaling here as well.

```python
import multilevel_tools

f_0 = numpy.ones( grid_0_size )
f_0 = array_tools.scaleVector( f_0, diag_0 )
    
f_1 = numpy.ones( grid_1_size )
f_1 = array_tools.scaleVector( f_1, diag_1 )

f_2 = numpy.ones( grid_2_size )
f_2 = array_tools.scaleVector( f_2, diag_2 )
```

## Monte Carlo Solution

Set the weight cutoff and number of histories to compute on each grid.

```python
w_c = 1.0e-2
np = 1000

np_ml = 4000
np_0 = int(np_ml*2**(-6.0/2.0))
np_1 = int(np_ml*2**(-3.0/2.0))
np_2 = int(np_ml)
```

Solve the problem with standard Monte Carlo

```python
import time
import mc_tools

start = time.time()
x, sigma = mc_tools.monteCarloSolve( A_0, f_0, w_c, np)
end = time.time()
time_x = end - start
print "1-level time:", time_x
```

Now compute the multilevel solution using Giles' estimator

```python
# Solve the levels
start = time.time()
x_0, sigma_0 = mc_tools.multilevelMonteCarloSolve( A_0, f_0, w_c, np_0 )
end = time.time()
time_0 = end - start

start = time.time()
x_1, sigma_1 = mc_tools.multilevelMonteCarloSolve( A_1, f_1, w_c, np_1 )
end = time.time()
time_1 = end - start

start = time.time()
x_2, sigma_2 = mc_tools.monteCarloSolve( A_2, f_2, w_c, np_2 )
end = time.time()
time_2 = end - start

time_ml = time_0 + time_1 + time_2
print "3-level time:", time_ml

# Combine the solution results
Px_2 = multilevel_tools.prolongateCC( x_2 )
for i in xrange(grid_1_size):
    x_1[i] = x_1[i] + Px_2[i]
    
x_ml = numpy.zeros(grid_0_size)
Px_1 = multilevel_tools.prolongateCC( x_1 )
for i in xrange(grid_0_size):
    x_ml[i] = Px_1[i] + x_0[i]
     
# Combine the variance results
Psigma_2 = multilevel_tools.prolongateCC( sigma_2 )
for i in xrange(grid_1_size):
    sigma_1[i] = sigma_1[i] + Psigma_2[i]
    
sigma_ml = numpy.zeros(grid_0_size)
Psigma_1 = multilevel_tools.prolongateCC( sigma_1 )
for i in xrange(grid_0_size):
    sigma_ml[i] = Psigma_1[i] + sigma_0[i]
```

## Reference Solution

```python
ref_0 = numpy.linalg.solve(A_0,f_0)
```

## Analysis

Compute the solution residual and error norms

```python
r = array_tools.computeResidual(A_0,x,f_0)
r_ml = array_tools.computeResidual(A_0,x_ml,f_0)

r_norm = numpy.linalg.norm(r,2)
r_norm_ml = numpy.linalg.norm(r_ml,2)

error = array_tools.relativeDiff(x,ref_0)
error_ml = array_tools.relativeDiff(x_ml,ref_0)

error_norm = numpy.linalg.norm(error,2)
error_norm_ml = numpy.linalg.norm(error_ml,2)

var_norm = numpy.linalg.norm(sigma,2)
var_norm_ml = numpy.linalg.norm(sigma_ml,2)

print "residual 2-norm:", r_norm, r_norm_ml
print "error 2-norm:", error_norm, error_norm_ml
print "variance 2-norm:", var_norm, var_norm_ml
```

Compute the figures of merit

```python
rfom = 1.0/(r_norm*r_norm*time_x)
rfom_ml = 1.0/(r_norm_ml*r_norm_ml*time_ml)
efom = 1.0/(error_norm*error_norm*time_x)
efom_ml = 1.0/(error_norm_ml*error_norm_ml*time_ml)
varfom = 1.0/(var_norm*time_x)
varfom_ml = 1.0/(var_norm_ml*time_ml)

print "RFOM", rfom, rfom_ml
print "EFOM:", efom, efom_ml
print "VFOM", varfom, varfom_ml
```
